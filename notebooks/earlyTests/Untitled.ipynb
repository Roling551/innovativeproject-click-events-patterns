```python
import ipywidgets as wg
from IPython.display import display
import pandas
import numpy as np
import bokeh
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
import functools
```

```python
def roundDown(number, divisor):
    return np.floor(number/divisor)*divisor;
```

```python
data = pandas.read_csv('C:\\Users\\Robert\\Desktop\\write.csv')
```

```python
data['time']=data.groupby('idVisit').cumsum()['timeSpent']
data["timeInterval"]=data.loc[:,'time'].apply(lambda x: np.floor(x/100))
data2 = data[data.timeInterval<1000]
s = data2.groupby('timeInterval').count()

id= [x for x in range (len(s))]
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')
p.line(id,s.time, legend="Temp.", line_width=2)
show(p)
#output_notebook() 
```
