```python
mostClickerUsers = data.groupby("country").size().sort_values().tail(20)
id = [x for x in range (len(mostClickerUsers))]
plt.barh(y=id,width=mostClickerUsers.tolist())
plt.yticks(id,mostClickerUsers.axes[0].values)
plt.savefig('mostClickerUsers.pdf', bbox_inches='tight')
```

```python
plt.figure(figsize=(20,10))
mostClickerUsers = data.groupby("organization").size().sort_values().tail(30)
id = [x for x in range (len(mostClickerUsers))]
plt.barh(y=id,width=mostClickerUsers.tolist())
plt.yticks(id,mostClickerUsers.axes[0].values)
plt.savefig('mostClickerUsers.pdf', bbox_inches='tight')
```
